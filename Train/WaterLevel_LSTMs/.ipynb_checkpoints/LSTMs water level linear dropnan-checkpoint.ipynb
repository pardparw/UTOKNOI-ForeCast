```python
import pandas as pd
import matplotlib.pyplot as plt
```

ref source : https://www.kaggle.com/datasets/konivat/three-gorges-dam-water-data

```python
df = pd.read_csv('threegorges-water-storage.csv')
df.tail()
```

```python
df.describe()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_test = df.measurement_date
```

```python
time_ls = []
for d in df.measurement_date:
    month_num = months.index(d.split()[1])+1
    time_format = d.split()
    time_format[1] = str(month_num)
    y, m, d = time_format[0], time_format[1], time_format[2]
    time_format = f"{d}-{m}-{y}"
    time_ls.append(time_format)
    
d_time = pd.DataFrame({'Date_time' : time_ls})
```

```python
# concat a clean date_time
df = pd.concat([d_time, df], axis = 1)
df = df.drop(columns='measurement_date')
df
```

```python
# set date_time as index
df.index = pd.to_datetime(df['Date_time'], format="%d-%m-%Y")
df[:2]
```

```python
water_level = df.upstream_water_level
train_valid_test_ratio = [0.7, 0.2, 0.1]
def split_size(n : float):
    return int(len(water_level) * n)
train_ratio = split_size(train_valid_test_ratio[0])
valid_ratio = split_size(train_valid_test_ratio[1])
test_ratio = split_size(train_valid_test_ratio[2])

train_series = water_level[:train_ratio]
valid_series = water_level[train_ratio:train_ratio+valid_ratio]
test_series = water_level[train_ratio+valid_ratio:train_ratio+valid_ratio+test_ratio]
len(train_series), len(valid_series), len(test_series)
```

```python
plt.plot(train_series)
plt.title('actual training series')
plt.plot()
```

```python
fill_train = train_series.dropna()
plt.plot(fill_train)
plt.title(' training series with dropna')
plt.plot()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].set_title('Actual Validation Series')
axs[0].plot(valid_series)
axs[1].plot(test_series)
axs[1].set_title('Actual Test series')
plt.show()
```

```python
drop_valid = valid_series.dropna()
drop_test = test_series.dropna()
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].set_title('Dropna Validation Series')
axs[0].plot(drop_valid)
axs[1].plot(drop_test)
axs[1].set_title('Dropna Test series')
plt.show()
```

```python
# replace old series with pre-processing datasets
train_series = fill_train
valid_series = drop_valid
test_series = drop_test
len(train_series), len(valid_series), len(test_series)
```

```python
import numpy as np
# preparing a data 
def df_to_X_y(df, window_size):
    df_as_np = df.to_numpy()
    X = []
    y = []
    for i in range(len(df_as_np)-window_size): 
        # just make sure there are enough data points remain
        row = [[a] for a in df_as_np[i:i+window_size]]
        X.append(row)
        y.append(df_as_np[i+window_size])
    return np.array(X), np.array(y)
```

```python
window_size = 5
X_train, y_train = df_to_X_y(df=train_series, window_size=window_size)
X_val, y_val = df_to_X_y(df=valid_series, window_size=window_size)
X_test, y_test = df_to_X_y(df=test_series, window_size=window_size)
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
classes = [X_train, y_train, X_val, y_val, X_test, y_test]
for c in classes:
    c = c.astype(np.float32)
    print(c.dtype is np.dtype(np.float32))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
model1 = Sequential([
    InputLayer((5,1)),
    LSTM(64),
    BatchNormalization(),
    Dense(8, activation='relu'),
    Dense(1, activation='linear')
])

model1.summary()
```

```python
import shutil
shutil.rmtree('LSTM_model_dropna/')
```

```python
cp = ModelCheckpoint('LSTM_model_dropna/', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
epochs = 600
history = model1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, callbacks=[cp])
```

```python
import matplotlib.pyplot as plt

# Create subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot train_loss and validation_loss on the left subplot
axs[0].set_title('Train Loss and Validation Loss')
axs[0].plot(history.history['loss'], label='Train Loss')
axs[0].plot(history.history['val_loss'], label='Validation Loss')
axs[0].set_ylabel('Loss')
axs[0].set_xlabel('Epochs')
axs[0].legend()

# Plot train_RMSE and val_RMSE on the right subplot
axs[1].set_title('Train RMSE and Validation RMSE')
axs[1].plot(history.history['root_mean_squared_error'], label='Train RMSE')
axs[1].plot(history.history['val_root_mean_squared_error'], label='Validation RMSE')
axs[1].set_ylabel('RMSE')
axs[1].set_xlabel('Epochs')
axs[1].legend()

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
classes = ['loss', 'val_loss', 'root_mean_squared_error', 'val_root_mean_squared_error']
for name in classes:
    print(f"{name} : {history.history[name][-1]:.4f}")
```

```python
# load a model
from tensorflow.keras.models import load_model
model1 = load_model('LSTM_model_dropna/')
```
